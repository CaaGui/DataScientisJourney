```python
import numpy as np

import statsmodels.formula.api as smf
import statsmodels.api as sm
import patsy
```

```python
df['data_ref'].value_counts()
```

```python
y , x = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos +\
                        C(tipo_renda,Treatment(0)) + C(educacao,Treatment(2)) + C(estado_civil,Treatment(0)) +\
                        C(tipo_residencia,Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia',data= df)
```

```python
reg = sm.OLS(y,x).fit().summary()
reg
```

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel +\
                idade + tempo_emprego',data= df).fit()

reg.summary()
```

```python
A cada variável removida o valor AIC apresentado foram menores do que o modelo inicial,
o valor do R2 diminui e do Adj. R2 também diminuiu.
Seguindo como parâmetro de rejeição as variáveis com valores P superiores a 5%
Acredito que mesmo não sendo o modelo com maior R2, um modelo mais simples
seja o melhor modelo.
```
