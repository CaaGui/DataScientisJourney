## Criando variáveis

**Variáveis temporais**: É possível que haja alguma regra de liberação de créditos de acordo com o dia do mês, o dia da semana, fim de semana (sim ou não), ou até alterações temporais da regra. Vamos criar as seguintes variáveis para testar estas hipóteses:

- Dia da semana
- Dummy para fim de semana
- Dia do mês
- Mês (já na base)
- Trimestre
- Ano (já na base)

**Valor da nota**: Vamos criar também algumas variáveis para resumir o valor da nota, e posteriormente criar uma intuição sobre categorizações de variáveis contínuas.

- Crie categorizações da variável ```Valor NF```, por quantís, em 5, 10, 20 e 50 categorias.

````python
df_ = df.dropna().copy()
````

````python
df_.loc[df_['Retorno']>0,['Retorno']]= 1
df_.loc[df_['Retorno']==0,['Retorno']]= 0
````

````python
df_['week_day'] = df_['Data Emissão'].dt.strftime("%u").astype('int64')
df_['week_end'] = df_['Data Emissão'].dt.strftime("%u").astype('int64')
df_['month_day'] = df_['Data Emissão'].dt.strftime("%d").astype('int64')
df_['month'] = df_['Data Emissão'].dt.strftime("%m").astype('int64')

df_['quarter'] = df_['Data Emissão'].dt.quarter
z = df_.set_index(['Ano','quarter']).index.to_list()
df_['quarter'] = ["%s-%02d" % item for item in z]
     
fil = (df_['week_end'] != 7) & (df_['week_end'] != 6)
df_.loc[fil,'week_end'] = 0
df_.loc[df_['week_end'] != 0,'week_end'] = 1
````

````python
df_['Q5'] = pd.qcut(df_['Valor NF'], 5, duplicates='drop')
df_['Q10'] = pd.qcut(df_['Valor NF'], 10, duplicates='drop')
df_['Q20'] = pd.qcut(df_['Valor NF'], 20, duplicates='drop')
df_['Q50'] = pd.qcut(df_['Valor NF'], 50, duplicates='drop')

````

## Calcule o IV de todas essas variáveis

Sugestão: Construa um metadados e tente utilizar uma função como a que foi utilizada durante a aula. Você vai conseguir fazer esse cálculo para as variáveis uma a uma, mas em casos reais em que vamos mais além na criação de variáveis, podemos ter centenas, ou até milhares de variáveis, esses cálculos *precisam* ser automáticos nesses casos.

````python
metadados = pd.DataFrame({'dtype':df_.dtypes})
metadados['papel'] = 'covariavel'
metadados.loc['Retorno','papel'] = 'resposta'
metadados['nunique'] = df_.nunique()
metadados = metadados.sort_values('papel',ascending=False)
````

````python
def IV(var_ex, var_endog):
    
    tab = pd.crosstab(var_ex, var_endog, margins=True, margins_name='Total')

    evento_0 = tab.columns[0]
    evento_1 = tab.columns[1]
 
    tab['evento0'] = tab[evento_0] / tab.loc['Total',evento_0]
    tab['evento1'] = tab[evento_1] / tab.loc['Total',evento_1]
    
    tab['woe'] = np.log(tab['evento1'] / tab['evento0'])
    
    tab['iv_parcial'] = (tab['evento1'] - tab['evento0']) * tab.woe
    
    return tab['iv_parcial'].sum()
````

````python
for var in metadados[metadados.papel=='covariavel'].index:
    metadados.loc[var, 'IV'] = IV(df_[var], df_.Retorno)
        
metadados.sort_values('IV')

````

## Conclua

- Quais variáveis parecem apresentar maior poder de discriminação segundo o IV?
- Quando categorizamos variáveis contínuas parece haver alguma relação entre o número de categorias e o IV?

# As variáveis com os melhores valores são, categoria com IV forte e os quantis, Q50 e Q20, com IV médio.
# Sim, parece haver uma relação com o número de variáveis e o valor de IV, quanto maior o número de variáveis maior o valor de IV.

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da variável ```categoria```.

````python
df_group = df_.groupby(['categoria','quarter'])[['Créditos']].sum().reset_index(level=0)
df_group = df_group.reset_index()
df_group
````

````python
meta = pd.DataFrame({'Categoria':df_.categoria.unique()})
````

````python
for cat in meta['Categoria']:
    filtro = (df_['categoria'] == cat)
    df_cat = df_[filtro]
    tab = pd.crosstab(df_cat['quarter'],df_cat['Créditos'], margins=True, margins_name= 'Total')

    evento_0 = tab.columns[0]
    evento_1 = tab.columns[1]

    tab['evento0'] = tab[evento_0] / tab.loc['Total',evento_0]
    tab['evento1'] = tab[evento_1] / tab.loc['Total',evento_1]

    tab['woe'] = np.log(tab['evento1'] / tab['evento0'])

    tab['iv_parcial'] = (tab['evento1'] - tab['evento0']) * tab.woe
    
    meta.loc[meta['Categoria']==cat, 'IV'] = tab['iv_parcial'].sum()
    
meta.sort_values('IV')
````
