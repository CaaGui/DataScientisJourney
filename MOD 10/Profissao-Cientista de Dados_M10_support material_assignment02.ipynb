````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
````

````python
sns.pointplot(data=df, x='posse_de_imovel', y='renda', ci=95)
````

````python
sns.pointplot(data=df, x='posse_de_veiculo', y='renda', ci=95)
````

````python
sns.pointplot(data=df, x='posse_de_imovel', y='renda', hue='posse_de_veiculo', ci=95)
````
