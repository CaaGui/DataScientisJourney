### 1. Crie 2 faixas de Latitude do município (munResLat) sendo uma acima e outra abaixo de -10.5 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
i=0
for lat in sinasc_raw.munResLat:
    if lat > (-10.5):
        sinasc_raw.loc[i,'LAT_FX']  = 'Acima'
    elif lat < (-10.5):
        sinasc_raw.loc[i,'LAT_FX']  = 'Abaixo'
    i += 1

sinasc_raw.groupby('LAT_FX')['QTDPARTCES'].agg(['sum','max','min','mean','median','std','var'])
```

### 2. Crie 2 faixas da área dos municípios (munResArea) sendo uma acima e outra abaixo de 3000 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas


```python
i=0
for area in sinasc_raw.munResArea:
    if area > 3000:
        sinasc_raw.loc[i,'ResArea_3000']  = 'Acima'
    elif area < 3000:
        sinasc_raw.loc[i,'ResArea_3000']  = 'Abaixo'
    i += 1
    
sinasc_raw.groupby(['ResArea_3000'])[['QTDGESTANT','QTDPARTNOR']].agg(['sum','max','min','mean','median','std','var'])
```

### 3. Determine faixas na variável munResAlt e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
sinasc_raw['munResAlt'].describe()
```

```python
i=0
for alt in sinasc_raw.munResAlt:
    if alt > 153:
        sinasc_raw.loc[i,'ResAlt_153']  = 'Acima'
    else:
        sinasc_raw.loc[i,'ResAlt_153']  = 'Abaixo'
    i += 1

sinasc_raw.groupby(['ResAlt_153'])[['QTDGESTANT','QTDPARTNOR']].agg(['sum','max','min','mean','median','std','var'])
```

### 5. Utilize a tabela do link abaixo e crie faixas utilizando o mapping e gere agrupamentos utilizando essas faixas como chave


### 5.1 IDH
A - https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IDH-M


```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IDH-M'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
table = soup.find('table',{'class':'wikitable sortable'}).tbody
rows = table.find_all('tr')
columns = [v.text.replace('\n','') for v in rows[0].find_all('th') + rows[1].find_all('th')]
columns.pop(2)

val =[]
for i in range(1,len(rows)):
  tds = rows[i].find_all('td')

  if len(tds) == 6:
    val.append({columns[0]:tds[0].text.replace('\n',''),
                columns[1]:tds[1].text.replace('\n',''),
                columns[2]:tds[2].text.replace('\n',''),
                columns[3]:tds[3].text.replace('\n',''),
                columns[4]:tds[4].text.replace('\n',''),
                columns[5]:tds[5].text.replace('\n','')
                }
               )

df_idh = pd.DataFrame(val)
```

### 5.2 IFDM 
B - https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IFDM


```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_IFDM'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
table = soup.find('table',{'class':'wikitable'}).tbody
rows = table.find_all('tr')
columns = [v.text.replace('\n','') for v in rows[0].find_all('th')]

val =[]
for i in range(1,len(rows)):
  tds = rows[i].find_all('td')

  if len(tds) == 3:
    val.append({columns[0]:tds[0].text.replace('\n',''),
                columns[1]:tds[1].text.replace('\n',''),
                columns[2]:tds[2].text.replace('\n','')
                }
               )

df_ifdm = pd.DataFrame(val)
```

### 5.3 PIB
C - https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_PIB


```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Rond%C3%B4nia_por_PIB'

r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
table = [tb.tbody for tb in soup.find_all('table',{'class':'wikitable'})]
rows = [tb.find_all('tr') for tb in table]
columns = [v.text.replace('\n','') for v in rows[0][0].find_all('th')]
columns.pop(3)

val =[]
for i in range(len(rows)):
  for j in range(2,len(rows[i])):
    tds = rows[i][j].find_all('td')

    if len(tds) == 5:
      val.append({columns[0]:tds[0].text.replace('\n',''),
                  columns[1]:tds[2].find('a').text.replace('\n',''),
                  columns[2]:tds[3].text.replace('\n','')
                  }
                )

df_pib = pd.DataFrame(val)
```
